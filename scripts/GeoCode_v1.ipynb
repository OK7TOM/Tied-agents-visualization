```python
import pandas as pd
import folium
from geopy.geocoders import Nominatim
from geopandas. tools import geocode
```

```python
#, sheet_name='MngFeeSAP', usecols='F,H,L,M:R'
df_input = pd.read_excel(r'_InputFiles/GeoSource_2021-12-31.xlsx', sheet_name='GeoImportNotFound', converters={'E-Icn':str})
df_input
```

```python
#, addressdetails=True, namedetails=True
geolocator = Nominatim(user_agent="mudrtomoch", timeout=10)
```

```python
def find_location(row):
    
    place = row['place']
    
    location = geolocator.geocode(place)
    
    if location != None:
        return location.address, location.latitude, location.longitude, location.raw['importance']
    else:
        return "Not Found", "Not Found", "Not Found", "Not Found"
```

```python
df_input[['address', 'lat', 'lon', 'importance']] = df_input.apply(find_location, axis="columns", result_type="expand")
df_input
```

```python
# Saving the dataframe to an excel file 
df_input.to_excel('GeoVZ2021-12-31NotFound.xlsx')      
```
